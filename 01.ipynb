```python
from sympy import *
# 1. Найти координаты точек пересечения
# 2. Построить графики функций в одной системе координат
# 3. Построить график функции пересечения
x = Symbol('x')
f = Function('f')
g = Function('g')
f = collect((x**3 - 50*x), x)
g = collect((-x**4 + 88*x**2 - 241), x)
c = g - f

z = solve(Eq(c, 0), x)

for i in z:
    print(i)
    print()
```

```python
for i in z:
    print()
    print(f.subs(x, i))
```
